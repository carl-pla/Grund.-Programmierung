## Abstraktion & Klassen 

Die Klasse `Exam` modelliert eine reale Prüfung mit den Datenattributen examiner, subject, course_number, participants und results.

```python
class Exam:
    def __init__(self, examiner,subject, course_number):
        self.examiner = examiner
        self.subject = subject
        self.course_number = course_number
        self.participants = [] #wissen wir noch nicht im Vorhinein, resultiert erst danacg
        self.results = [] #hier das selbe
    def get_number_of_participants(self):
        return len(self.participants)
```

#### Erzeugen einer neuen Prüfungsinstanz für die Analysis-Prüfung:

```python
analysis_exam = Exam("GJ","Analysis","WDS25B1")
```

#### Zugriff auf die Attribute mittels dot-Notation

```python
analysis_exam.examiner
```

```python
analysis_exam.get_number_of_participants()
```

#### Hinzufügen von Personen, die an der Prüfung teilnehmen:

```python
analysis_exam.participants = [
    "Lovelace, Ada",
    "Turing, Alan",
    "Hopper, Grace",
    "von Neumann, John",
    "Knuth, Donald"
]
```

```python
analysis_exam.participants.append("Placek, Carl")
analysis_exam.participants.remove("Placek, Carl")
print(analysis_exam.participants)
```

Über die Methode `get_number_of_participants()` der Schnittstelle (Interface) der Klasse `exam` kann die Anzahl der Teilnehmenden ermittelt werden, um z.B. einen geeigneten Raum für die Prüfung zu ermitteln.

```python
analysis_exam.get_number_of_participants()
```

#### Analog kann eine weitere Instanz erzeugt werden

```python
la_exam = Exam("AW","Lineare Algebra","WDS25B1")
```

Hierdurch entstehen unterschiedliche Objekte desselben Typs

```python
la_exam is analysis_exam
```

```python
type(analysis_exam) == type(la_exam)
```

```python
print(type(la_exam))
```

Sie können diese Objekte auch in Listen, Tupeln, etc. verwenden:

```python
exams = [analysis_exam, la_exam]
for exam in exams:
    print(f"{exam.subject}: {exam.get_number_of_participants()}")
```

### Erweiterung der Funktionalität

Falls nun einzelne Teilnehmende zusätzlich einer Prüfung zugeordnet werden sollen, könnte man z.B. eine neue Methode ergänzen:

```python
class Exam:
    def __init__(self, examiner,subject, course_number):
        self.examiner = examiner
        self.subject = subject
        self.course_number = course_number
        self.participants = []
        self.results = []
    def get_number_of_participants(self):
        return len(self.participants)
    def add_participant(self, participant):
        self.participants.append(participant)
```

```python
la_exam = Exam("AW","Lineare Algebra","WDS25B1")
```

```python
la_exam.add_participant("van Rossum, Guido")
```

```python
print(la_exam.participants)
```

```python
la_exam.participants.append("van Rossum, Guido")
```

Man könnte nun argumentieren, dass der Aufruf  
`la_exam.add_participant("van Rossum, Guido")`  
im Vergleich zum äquivalenten Aufruf  
`la_exam.participants.append("van Rossum, Guido")`  
nicht kürzer ist.

Allerdings, und darum geht es bei der Abstraktion, bietet die Methode  
`add_participant`   
eine Schnittstelle (Interface) für das Hinzufügen eines Teilnehmenden
ohne dabei genaueres über die interne Datenstruktur des Attributs `participants` wissen zu müssen.
