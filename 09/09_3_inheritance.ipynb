## Vererbung

```python
class Exam:
    def __init__(self, examiner, subject, course_number):
        self.examiner = examiner
        self.subject = subject
        self.course_number = course_number
        self.participants = []
        self.results = []

    def get_number_of_participants(self):
        return len(self.participants)

    def add_participant(self, participant):
        self.participants.append(participant)
```

Wir leiten nun von der Klasse `Exam`eine Spezialisierung `WrittenExam` ab.  
Die Klasse `WrittenExam` erbt dann alle Attribute von `Exam`und fügt der Klasse weitere hinzu.  
Mit `super()` kann hierbei auf die übergeordnete Klasse `Exam`zugegriffen werden.

```python
class WrittenExam(Exam):
    TYPE = "Klausur"  # Art der Prüfung als Konstante der Klasse

    def __init__(self, examiner, subject, course_number, date, time, duration):
        super().__init__(examiner, subject, course_number)
        self.date = date
        self.time = time
        self.duration = duration
```

#### Wir legen zwei Klausuren im Klausurplanungssystem an:

```python
analysis_klausur = WrittenExam(
    examiner="GJ",
    subject="Analysis",
    course_number="WDS24B1",
    date="18.12.2024",
    time="10:00",
    duration="60 min"
)

la_klausur = WrittenExam(
    examiner="AW",
    subject="Lineare Algebra",
    course_number="WDS24B1",
    date="18.12.2024",
    time="10:00",
    duration="60 min"
)
```

```python
analysis_klausur.add_participant("Yoda")
analysis_klausur.add_participant("Luke_Skywalker")
la_klausur.add_participant("Yoda")

exams = [analysis_klausur, la_klausur]
```

#### Die Methode get_number_of_participants() kann jeweils gerufen werden, da die Kind-Klassen diese Methode von der Exam-Klasse geerbt haben.

```python
print(f"Analysis: Anzahl Teilnehmende {analysis_klausur.get_number_of_participants()} ")
```

```python
for exam in exams:
    print(f"{exam.TYPE} in {exam.subject}, Anzahl Teilnehmende: {exam.get_number_of_participants()}")
```
