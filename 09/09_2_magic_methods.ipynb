### Magic Methods: `__add__`

#### Wird die Methode `__add__` in einer Klasse definiert, dann können Instanzen der Klasse mittels des Operators `+` verknüpft werden.  Bei dieser Verknüpfung wird dann die Methode `__add__` für die Instanz gerufen.

(Namenskonvention bei Klassennamen: CamelCase)

```python
class MyTuple3d:
    def __init__(self,x,y,z):
        self.x = x
        self.y = y
        self.z = z
    def __add__(self,summand2):
        result = MyTuple3d(
            self.x + summand2.x,
            self.y + summand2.y,
            self.z + summand2.z
        )
        return result
```

```python
u = MyTuple3d(1,2,3)
v = MyTuple3d(4,5,6)
summe = u+v
print(summe.x, summe.y, summe.z)
```

Dies ist äquivalent zu:

```python
u = MyTuple3d(1,2,3)
v = MyTuple3d(4,5,6)
summe = u.__add__(v)
print(summe.x, summe.y, summe.z)
```
