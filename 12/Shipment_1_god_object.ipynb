### Modellierung von Versand

Modellierung der Versandarten
- Standard
- Express

mit Berechnung der Versandkosten.

```python
class Shipment:
    def __init__(self,type,weight):
        self.type = type
        self.weight = weight
    def calculate_cost(self):
        if self.type == "standard":
            if self.weight <= 100:
                return 5
            else:
                return 10
        elif self.type == "express":
            if self.weight <= 100:
                return 15
            else:
                return 30
```

```python
shipment_1 = Shipment("standard",10)
print(shipment_1.calculate_cost())
```

Hier wird bei der Klassen-Modellierung nicht zwischen den verschiedenen Lieferarten unterschieden. Es gibt somit nur eine Klasse mit einigen `if`-Abfragen.
Bei Erweiterungen (bei neuen zusätzlichen Lieferarten) muss die Methode `calculate_cost` angepasst werden, was diese mit jeder Erweiterung größer und größer werden lässt.
Damit leidet die Wartbarkeit und ab einem gewissen Zeitpunkt versteht niemand mehr, was die Methode genau macht.

Solche Klassen, die sehr viel unterschiedliche Funktionalität beinhalten, werden auch God Object genannt.
